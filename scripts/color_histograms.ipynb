```python
import os.path
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Open the camera
#camera= cv2.VideoCapture(0)
PATH_1 = '/home/marilin/Documents/ESP/data/SYTO_PI/control_50killed_syto_PI_2-Image Export-02_c1-3.jpg'
img = cv2.imread(PATH_1,0)

plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
img = cv2.imread(PATH_1)
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
